```python
import numpy as np

# 1. Filter the df_phq9 DataFrame to create a new column, 'suicidal_ideation_target'
#    Map 'Not at all' to 0, and all other values to 1.
#    Ensure this new column is of integer type.
df_phq9['suicidal_ideation_target'] = np.where(
    df_phq9['thoughts_death'] == 'Not at all', 0, 1
).astype(int)

# 2. Create a new DataFrame, df_suicidal_ideation_target, containing only 'participant_id' and the new 'suicidal_ideation_target' column.
df_suicidal_ideation_target = df_phq9[['participant_id', 'suicidal_ideation_target']].copy()

print("df_phq9 with new 'suicidal_ideation_target' column created successfully.")
print("df_suicidal_ideation_target created successfully. Displaying head and shape:")
print(df_suicidal_ideation_target.head())
print(f"Shape of df_suicidal_ideation_target: {df_suicidal_ideation_target.shape}")
```

```python
# 3. Merge df_suicidal_ideation_target with the existing df_features DataFrame
#    using 'participant_id' as the key. Store the result in a new DataFrame, df_merged_data_suicide.
df_merged_data_suicide = pd.merge(
    df_features,
    df_suicidal_ideation_target,
    on='participant_id',
    how='left' # Use a left merge to keep all entries from df_features
)

# 4. Display the first 5 rows and the shape of the newly merged DataFrame (df_merged_data_suicide)
print("df_merged_data_suicide created successfully. Displaying head and shape:")
print(df_merged_data_suicide.head())
print(f"Shape of df_merged_data_suicide: {df_merged_data_suicide.shape}")
```

```python
# Identify non-numeric columns to drop (excluding 'participant_id' which is needed for merging)
columns_to_drop_suicide = ['session_id', 'task_name', 'transcription']

# Drop the identified non-numeric columns from a copy of df_merged_data_suicide
df_processed_suicide = df_merged_data_suicide.drop(columns=columns_to_drop_suicide, errors='ignore').copy()

print(f"Columns dropped: {columns_to_drop_suicide}")
print("DataFrame after dropping irrelevant non-numeric columns. Displaying head and shape:")
print(df_processed_suicide.head())
print(f"Shape of df_processed_suicide: {df_processed_suicide.shape}")
```

```python
# Drop rows where the 'suicidal_ideation_target' column is NaN
df_processed_suicide.dropna(subset=['suicidal_ideation_target'], inplace=True)

print("DataFrame after dropping rows with NaN in 'suicidal_ideation_target' column. Displaying head and shape:")
print(df_processed_suicide.head())
print(f"Shape of df_processed_suicide: {df_processed_suicide.shape}")
```

```python
import numpy as np

# Convert 'suicidal_ideation_target' column to integer type
df_processed_suicide['suicidal_ideation_target'] = df_processed_suicide['suicidal_ideation_target'].astype(int)

# Identify numeric columns for imputation, excluding 'participant_id' and 'suicidal_ideation_target'
numeric_cols_suicide = df_processed_suicide.select_dtypes(include=np.number).columns.tolist()
if 'participant_id' in numeric_cols_suicide:
    numeric_cols_suicide.remove('participant_id')
if 'suicidal_ideation_target' in numeric_cols_suicide:
    numeric_cols_suicide.remove('suicidal_ideation_target')

# Impute missing values in the selected numeric columns with their respective means
for col in numeric_cols_suicide:
    if df_processed_suicide[col].isnull().any():
        df_processed_suicide[col] = df_processed_suicide[col].fillna(df_processed_suicide[col].mean())

print("Converted 'suicidal_ideation_target' to int and imputed missing numerical values using the mean.")
print("Displaying the first 5 rows of df_processed_suicide after imputation:")
print(df_processed_suicide.head())
print(f"Number of missing values after imputation: {df_processed_suicide[numeric_cols_suicide].isnull().sum().sum()}")
```

```python
from sklearn.model_selection import train_test_split

# 1. Separate the df_processed_suicide DataFrame into features (X_suicide) and the target variable (y_suicide)
X_suicide = df_processed_suicide.drop(columns=['participant_id', 'suicidal_ideation_target'])
y_suicide = df_processed_suicide['suicidal_ideation_target']

# 2. Split the features and target into training and testing sets (70/30 ratio)
X_train_suicide, X_test_suicide, y_train_suicide, y_test_suicide = train_test_split(X_suicide, y_suicide, test_size=0.3, random_state=42)

print("Data split into training and testing sets successfully.")
print(f"Shape of X_suicide: {X_suicide.shape}")
print(f"Shape of y_suicide: {y_suicide.shape}")
print(f"Shape of X_train_suicide: {X_train_suicide.shape}")
print(f"Shape of X_test_suicide: {X_test_suicide.shape}")
print(f"Shape of y_train_suicide: {y_train_suicide.shape}")
print(f"Shape of y_test_suicide: {y_test_suicide.shape}")
```

```python
from xgboost import XGBClassifier

# 1. Instantiate an XGBClassifier with specified parameters
#    eval_metric='logloss' and random_state=42 for reproducibility.
xgb_classifier_suicide = XGBClassifier(eval_metric='logloss', random_state=42)

# 2. Train the classifier using the training data for suicidal ideation
xgb_classifier_suicide.fit(X_train_suicide, y_train_suicide)

print("XGBoost Classifier for suicidal ideation initialized and trained successfully.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 1. Use the trained xgb_classifier_suicide to make class predictions on the X_test_suicide dataset.
#    Store the predicted classes as y_pred_suicide.
y_pred_suicide = xgb_classifier_suicide.predict(X_test_suicide)

# 2. Use the trained xgb_classifier_suicide to make probability predictions on the X_test_suicide dataset.
#    Extract the probabilities for the positive class (class 1) and store them as y_pred_proba_suicide.
y_pred_proba_suicide = xgb_classifier_suicide.predict_proba(X_test_suicide)[:, 1]

# 3. Calculate the accuracy score
accuracy_suicide = accuracy_score(y_test_suicide, y_pred_suicide)

# 4. Calculate the precision score
precision_suicide = precision_score(y_test_suicide, y_pred_suicide)

# 5. Calculate the recall score
recall_suicide = recall_score(y_test_suicide, y_pred_suicide)

# 6. Calculate the F1-score
f1_suicide = f1_score(y_test_suicide, y_pred_suicide)

# 7. Calculate the ROC AUC score
roc_auc_suicide = roc_auc_score(y_test_suicide, y_pred_proba_suicide)

# 8. Print all calculated evaluation metrics
print(f"Model Evaluation Metrics for Suicidal Ideation Prediction:")
print(f"Accuracy:  {accuracy_suicide:.4f}")
print(f"Precision: {precision_suicide:.4f}")
print(f"Recall:    {recall_suicide:.4f}")
print(f"F1-Score:  {f1_suicide:.4f}")
print(f"ROC AUC:   {roc_auc_suicide:.4f}")
```

```python
import pandas as pd

# 1. Access the feature_importances_ attribute
importance_scores_suicide = xgb_classifier_suicide.feature_importances_

# 2. Get the feature names from X_train_suicide.columns
feature_names_suicide = X_train_suicide.columns

# 3. Create a pandas DataFrame with 'Feature' and 'Importance'
df_feature_importances_suicide = pd.DataFrame({
    'Feature': feature_names_suicide,
    'Importance': importance_scores_suicide
})

# 4. Sort this DataFrame in ascending order based on the 'Importance' column
df_feature_importances_suicide_sorted = df_feature_importances_suicide.sort_values(by='Importance', ascending=True)

# 5. Print the sorted DataFrame
print("Feature Importances for Suicidal Ideation Prediction (sorted in ascending order):")
print(df_feature_importances_suicide_sorted)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
import numpy as np

# 1. Initialize two empty lists
num_features_list_suicide = []
roc_auc_scores_list_suicide = []

# Get all feature names from X_train_suicide
all_features_suicide = X_train_suicide.columns.tolist()

# Get features sorted by importance (ascending)
sorted_features_by_importance_suicide = df_feature_importances_suicide_sorted['Feature'].tolist()

# Loop through features, starting with all and removing the least important one by one
for i in range(len(sorted_features_by_importance_suicide) + 1):
    if i == 0: # Start with all features
        current_features_suicide = all_features_suicide
    else:
        # Remove the least important features iteratively
        current_features_suicide = [f for f in all_features_suicide if f not in sorted_features_by_importance_suicide[:i]]

    # Ensure there are features to train with
    if not current_features_suicide:
        continue

    # 3. Create reduced training and testing sets
    X_train_suicide_reduced = X_train_suicide[current_features_suicide]
    X_test_suicide_reduced = X_test_suicide[current_features_suicide]

    # 4. Initialize a new XGBClassifier
    xgb_classifier_suicide_reduced = XGBClassifier(eval_metric='logloss', random_state=42)

    # 5. Train this new classifier
    xgb_classifier_suicide_reduced.fit(X_train_suicide_reduced, y_train_suicide)

    # 6. Make probability predictions on X_test_suicide_reduced
    y_pred_proba_suicide_reduced = xgb_classifier_suicide_reduced.predict_proba(X_test_suicide_reduced)[:, 1]

    # 7. Calculate the ROC AUC score
    roc_auc_suicide_reduced = roc_auc_score(y_test_suicide, y_pred_proba_suicide_reduced)

    # 8. Append the current number of features and the calculated ROC AUC score
    num_features_list_suicide.append(len(current_features_suicide))
    roc_auc_scores_list_suicide.append(roc_auc_suicide_reduced)

    print(f"Iteration {i}: Features = {len(current_features_suicide)}, ROC AUC = {roc_auc_suicide_reduced:.4f}")


# Print the lengths of the lists to confirm they contain data.
print(f"\nLength of num_features_list_suicide: {len(num_features_list_suicide)}")
print(f"Length of roc_auc_scores_list_suicide: {len(roc_auc_scores_list_suicide)}")
```

```python
import pandas as pd
import numpy as np

# 1. Filter the df_processed_suicide DataFrame to get all rows where suicidal_ideation_target is 1,
#    and extract their unique participant_id's into a variable named participant_ids_class_1.
participant_ids_class_1 = df_processed_suicide[df_processed_suicide['suicidal_ideation_target'] == 1]['participant_id'].unique()

# 2. Filter the df_processed_suicide DataFrame to get all rows where suicidal_ideation_target is 0,
#    and extract their unique participant_id's into a variable named participant_ids_class_0.
participant_ids_class_0 = df_processed_suicide[df_processed_suicide['suicidal_ideation_target'] == 0]['participant_id'].unique()

# 3. Determine the number of unique participant_id's in participant_ids_class_1.
#    This will be the target count for undersampling.
target_count = len(participant_ids_class_1)

# 4. Randomly select participant_id's from participant_ids_class_0 to match the target count from step 3.
#    Store these selected IDs in undersampled_participant_ids_class_0.
#    Use a random_state for reproducibility.
np.random.seed(42) # Set random state for reproducibility
undersampled_participant_ids_class_0 = np.random.choice(participant_ids_class_0, size=target_count, replace=False)

# 5. Combine participant_ids_class_1 and undersampled_participant_ids_class_0
#    into a single list or Series called balanced_participant_ids.
balanced_participant_ids = np.concatenate((participant_ids_class_1, undersampled_participant_ids_class_0))

# 6. Print the number of unique participant_id's in participant_ids_class_1 and
#    undersampled_participant_ids_class_0 to confirm that an equal number of participant IDs
#    has been selected for each class.
print(f"Number of unique participant IDs for suicidal ideation (Class 1): {len(participant_ids_class_1)}")
print(f"Number of unique participant IDs for undersampled no suicidal ideation (Class 0): {len(undersampled_participant_ids_class_0)}")
print(f"Total unique participant IDs in balanced set: {len(balanced_participant_ids)}")
```

```python
# 1. Filter the df_processed_suicide DataFrame to include only the rows where participant_id is present in the balanced_participant_ids array.
#    Store the result in a new DataFrame named df_undersampled_suicide.
df_undersampled_suicide = df_processed_suicide[df_processed_suicide['participant_id'].isin(balanced_participant_ids)].copy()

# 2. Display the first 5 rows and the shape of the df_undersampled_suicide DataFrame to verify the successful creation and filtering of the data.
print("df_undersampled_suicide created and filtered successfully. Displaying head and shape:")
print(df_undersampled_suicide.head())
print(f"Shape of df_undersampled_suicide: {df_undersampled_suicide.shape}")
```

```python
# 1. Create a new DataFrame named X_undersampled_suicide by dropping the 'participant_id' and 'suicidal_ideation_target' columns from df_undersampled_suicide.
X_undersampled_suicide = df_undersampled_suicide.drop(columns=['participant_id', 'suicidal_ideation_target'])

# 2. Create a new Series named y_undersampled_suicide by selecting the 'suicidal_ideation_target' column from df_undersampled_suicide.
y_undersampled_suicide = df_undersampled_suicide['suicidal_ideation_target']

# 3. Print the shape of X_undersampled_suicide and y_undersampled_suicide to verify the separation.
print(f"Shape of X_undersampled_suicide: {X_undersampled_suicide.shape}")
print(f"Shape of y_undersampled_suicide: {y_undersampled_suicide.shape}")

print("First 5 rows of X_undersampled_suicide:")
print(X_undersampled_suicide.head())
print("First 5 rows of y_undersampled_suicide:")
print(y_undersampled_suicide.head())
```

```python
from sklearn.model_selection import train_test_split

# 1. Split X_undersampled_suicide and y_undersampled_suicide into training and testing sets (70/30 ratio)
X_train_undersampled, X_test_undersampled, y_train_undersampled, y_test_undersampled = train_test_split(
    X_undersampled_suicide,
    y_undersampled_suicide,
    test_size=0.3,
    random_state=42
)

# 2. Print the shapes of the newly created training and testing sets to verify the split.
print("Undersampled data split into training and testing sets successfully.")
print(f"Shape of X_train_undersampled: {X_train_undersampled.shape}")
print(f"Shape of X_test_undersampled: {X_test_undersampled.shape}")
print(f"Shape of y_train_undersampled: {y_train_undersampled.shape}")
print(f"Shape of y_test_undersampled: {y_test_undersampled.shape}")
```

```python
from xgboost import XGBClassifier

# 1. Instantiate an XGBClassifier with specified parameters
#    eval_metric='logloss' and random_state=42 for reproducibility.
xgb_classifier_undersampled = XGBClassifier(eval_metric='logloss', random_state=42)

# 2. Train the classifier using the undersampled training data for suicidal ideation
xgb_classifier_undersampled.fit(X_train_undersampled, y_train_undersampled)

print("XGBoost Classifier for suicidal ideation (undersampled) initialized and trained successfully.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 1. Use the trained xgb_classifier_undersampled to make class predictions on the X_test_undersampled dataset.
#    Store the predicted classes as y_pred_undersampled.
y_pred_undersampled = xgb_classifier_undersampled.predict(X_test_undersampled)

# 2. Use the trained xgb_classifier_undersampled to make probability predictions on the X_test_undersampled dataset.
#    Extract the probabilities for the positive class (class 1) and store them as y_pred_proba_undersampled.
y_pred_proba_undersampled = xgb_classifier_undersampled.predict_proba(X_test_undersampled)[:, 1]

# 3. Calculate the accuracy score
accuracy_undersampled = accuracy_score(y_test_undersampled, y_pred_undersampled)

# 4. Calculate the precision score
precision_undersampled = precision_score(y_test_undersampled, y_pred_undersampled)

# 5. Calculate the recall score
recall_undersampled = recall_score(y_test_undersampled, y_pred_undersampled)

# 6. Calculate the F1-score
f1_undersampled = f1_score(y_test_undersampled, y_pred_undersampled)

# 7. Calculate the ROC AUC score
roc_auc_undersampled = roc_auc_score(y_test_undersampled, y_pred_proba_undersampled)

# 8. Print all calculated evaluation metrics
print(f"Model Evaluation Metrics for Undersampled Suicidal Ideation Prediction:")
print(f"Accuracy:  {accuracy_undersampled:.4f}")
print(f"Precision: {precision_undersampled:.4f}")
print(f"Recall:    {recall_undersampled:.4f}")
print(f"F1-Score:  {f1_undersampled:.4f}")
print(f"ROC AUC:   {roc_auc_undersampled:.4f}")
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import pandas as pd
import os

# --- Re-load and preprocess data to ensure df_processed_suicide and participant_ids are defined ---

# Define the full path to the `phq9.tsv` file and load it into df_phq9
phq9_file_path = os.path.join(
    '/content/drive/MyDrive/bridge2ai-voice-an-ethically-sourced-diverse-voice-dataset-linked-to-health-information-3.0.0/phenotype/questionnaire',
    'phq9.tsv'
)
df_phq9 = pd.read_csv(phq9_file_path, sep='\t')

# Create 'suicidal_ideation_target' column in df_phq9
df_phq9['suicidal_ideation_target'] = np.where(
    df_phq9['thoughts_death'] == 'Not at all', 0, 1
).astype(int)

# Define the full path to the `static_features.tsv` file and load it into df_features
features_directory_path = '/content/drive/MyDrive/bridge2ai-voice-an-ethically-sourced-diverse-voice-dataset-linked-to-health-information-3.0.0/features'
dataset_file_name = 'static_features.tsv'
full_dataset_path = os.path.join(features_directory_path, dataset_file_name)
df_features = pd.read_csv(full_dataset_path, sep='\t')

# Create df_suicidal_ideation_target
df_suicidal_ideation_target = df_phq9[['participant_id', 'suicidal_ideation_target']].copy()

# Merge df_features with df_suicidal_ideation_target
df_merged_data_suicide = pd.merge(
    df_features,
    df_suicidal_ideation_target,
    on='participant_id',
    how='left'
)

# Drop irrelevant non-numeric columns
columns_to_drop_suicide = ['session_id', 'task_name', 'transcription']
df_processed_suicide = df_merged_data_suicide.drop(columns=columns_to_drop_suicide, errors='ignore').copy()

# Drop rows where 'suicidal_ideation_target' is NaN
df_processed_suicide.dropna(subset=['suicidal_ideation_target'], inplace=True)

# Convert 'suicidal_ideation_target' to integer type
df_processed_suicide['suicidal_ideation_target'] = df_processed_suicide['suicidal_ideation_target'].astype(int)

# Impute missing numerical feature values with the mean
numeric_cols_suicide = df_processed_suicide.select_dtypes(include=np.number).columns.tolist()
if 'participant_id' in numeric_cols_suicide:
    numeric_cols_suicide.remove('participant_id')
if 'suicidal_ideation_target' in numeric_cols_suicide:
    numeric_cols_suicide.remove('suicidal_ideation_target')

for col in numeric_cols_suicide:
    if df_processed_suicide[col].isnull().any():
        df_processed_suicide[col] = df_processed_suicide[col].fillna(df_processed_suicide[col].mean())

# Get unique participant_ids for class 0 and class 1
participant_ids_class_1 = df_processed_suicide[df_processed_suicide['suicidal_ideation_target'] == 1]['participant_id'].unique()
participant_ids_class_0 = df_processed_suicide[df_processed_suicide['suicidal_ideation_target'] == 0]['participant_id'].unique()

# --- End of re-load and preprocess data ---

# Re-calculate top_52_features_suicide just in case it was not defined in the current kernel session
# This part assumes df_feature_importances_suicide_sorted has been computed and is available from previous cells.
# If not, you might need to re-run the cells that generate it (D-aftkAtazKn and ZhlgZuoHcxz-).
importance_scores_suicide = XGBClassifier(eval_metric='logloss', random_state=42).fit(
    df_processed_suicide.drop(columns=['participant_id', 'suicidal_ideation_target']),
    df_processed_suicide['suicidal_ideation_target']
).feature_importances_

df_feature_importances_suicide = pd.DataFrame({
    'Feature': df_processed_suicide.drop(columns=['participant_id', 'suicidal_ideation_target']).columns,
    'Importance': importance_scores_suicide
})
df_feature_importances_suicide_sorted = df_feature_importances_suicide.sort_values(by='Importance', ascending=True)
df_feature_importances_suicide_desc = df_feature_importances_suicide_sorted.sort_values(by='Importance', ascending=False)
top_52_df_suicide = df_feature_importances_suicide_desc.head(52)
top_52_features_suicide = top_52_df_suicide['Feature'].tolist()

# 1. Initialize empty lists to store the metrics for each bootstrap iteration
roc_auc_scores_bootstrap = []
accuracy_scores_bootstrap = []
precision_scores_bootstrap = []
recall_scores_bootstrap = []
f1_scores_bootstrap = []

# Determine the number of unique participant_id's in participant_ids_class_1.
# This will be the target count for undersampling in each bootstrap iteration.
target_count_bootstrap = len(participant_ids_class_1)

# 2. Start a loop that runs 100 times to perform the bootstrap resampling
for i in range(100):
    # a. Randomly sample target_count_bootstrap participant_id's with replacement from the 'no suicidal ideation' group (class 0).
    #    Use np.random.choice with replacement=True for bootstrapping.
    np.random.seed(i) # Set random seed for reproducibility of sampling in each iteration
    sampled_ids_class_0_bootstrap = np.random.choice(participant_ids_class_0, size=target_count_bootstrap, replace=True)

    # b. Combine these sampled participant_id's from class 0 with all participant_id's from class 1
    balanced_participant_ids_bootstrap = np.concatenate((participant_ids_class_1, sampled_ids_class_0_bootstrap))

    # c. Filter the df_processed_suicide DataFrame to create a new DataFrame, df_bootstrap_suicide
    df_bootstrap_suicide = df_processed_suicide[df_processed_suicide['participant_id'].isin(balanced_participant_ids_bootstrap)].copy()

    # d. Separate the features (X_bootstrap) and the target variable (y_bootstrap) from df_bootstrap_suicide.
    #    Filter X_bootstrap to use only the top 52 features.
    X_bootstrap = df_bootstrap_suicide[top_52_features_suicide]
    y_bootstrap = df_bootstrap_suicide['suicidal_ideation_target']

    # e. Split X_bootstrap and y_bootstrap into training and testing sets (70/30 ratio)
    X_train_bootstrap, X_test_bootstrap, y_train_bootstrap, y_test_bootstrap = train_test_split(
        X_bootstrap, y_bootstrap, test_size=0.3, random_state=42 # fixed random_state for consistency within bootstrap splits
    )

    # f. Initialize a new XGBClassifier
    xgb_classifier_bootstrap = XGBClassifier(eval_metric='logloss', random_state=42)

    # g. Train the classifier using the current iteration's training data
    xgb_classifier_bootstrap.fit(X_train_bootstrap, y_train_bootstrap)

    # h. Make class predictions (y_pred_bootstrap) and probability predictions (y_pred_proba_bootstrap)
    y_pred_bootstrap = xgb_classifier_bootstrap.predict(X_test_bootstrap)
    y_pred_proba_bootstrap = xgb_classifier_bootstrap.predict_proba(X_test_bootstrap)[:, 1]

    # i. Calculate the ROC AUC, accuracy, precision, recall, and F1-score
    roc_auc_current_iteration = roc_auc_score(y_test_bootstrap, y_pred_proba_bootstrap)
    accuracy_current_iteration = accuracy_score(y_test_bootstrap, y_pred_bootstrap)
    precision_current_iteration = precision_score(y_test_bootstrap, y_pred_bootstrap)
    recall_current_iteration = recall_score(y_test_bootstrap, y_pred_bootstrap)
    f1_current_iteration = f1_score(y_test_bootstrap, y_pred_bootstrap)

    # j. Append each calculated metric to its respective list
    roc_auc_scores_bootstrap.append(roc_auc_current_iteration)
    accuracy_scores_bootstrap.append(accuracy_current_iteration)
    precision_scores_bootstrap.append(precision_current_iteration)
    recall_scores_bootstrap.append(recall_current_iteration)
    f1_scores_bootstrap.append(f1_current_iteration)

    if (i + 1) % 10 == 0:
        print(f"Iteration {i+1}/100 complete. ROC AUC: {roc_auc_current_iteration:.4f}")

# 4. After the loop, print the lengths of all the lists
print(f"\nLength of ROC AUC scores list: {len(roc_auc_scores_bootstrap)}")
print(f"Length of Accuracy scores list: {len(accuracy_scores_bootstrap)}")
print(f"Length of Precision scores list: {len(precision_scores_bootstrap)}")
print(f"Length of Recall scores list: {len(recall_scores_bootstrap)}")
print(f"Length of F1-Score scores list: {len(f1_scores_bootstrap)}")
```

```python
# 1. Calculate the mean and standard deviation for each metric
mean_roc_auc = np.mean(roc_auc_scores_bootstrap)
std_roc_auc = np.std(roc_auc_scores_bootstrap)

mean_accuracy = np.mean(accuracy_scores_bootstrap)
std_accuracy = np.std(accuracy_scores_bootstrap)

mean_precision = np.mean(precision_scores_bootstrap)
std_precision = np.std(precision_scores_bootstrap)

mean_recall = np.mean(recall_scores_bootstrap)
std_recall = np.std(recall_scores_bootstrap)

mean_f1_score = np.mean(f1_scores_bootstrap)
std_f1_score = np.std(f1_scores_bootstrap)

# 2. Print the mean and standard deviation for each metric
print("\nSummary of Bootstrap Resampling Results (100 iterations):")
print(f"Mean ROC AUC:   {mean_roc_auc:.4f} (Std: {std_roc_auc:.4f})")
print(f"Mean Accuracy:  {mean_accuracy:.4f} (Std: {std_accuracy:.4f})")
print(f"Mean Precision: {mean_precision:.4f} (Std: {std_precision:.4f})")
print(f"Mean Recall:    {mean_recall:.4f} (Std: {std_recall:.4f})")
print(f"Mean F1-Score:  {mean_f1_score:.4f} (Std: {std_f1_score:.4f})")
```

3-fold cross validation approach for each of the 100 boostrapping iterations.



*   28 positive participants (same each time)
*   28 negative participants (sampled with replacement)





```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print("Preparing held-out test set...")
# Prepare the held-out test set once before the main bootstrap loop
X_held_out_test = df_held_out_test_suicide[top_52_features_suicide].copy()
y_held_out_test = df_held_out_test_suicide['suicidal_ideation_target'].copy()

print(f"Shape of X_held_out_test: {X_held_out_test.shape}")
print(f"Shape of y_held_out_test: {y_held_out_test.shape}")

# Initialize lists to store the mean metrics from each 3-fold CV in each bootstrap iteration
bootstrap_roc_auc_means = []
bootstrap_accuracy_means = []
bootstrap_precision_means = []
bootstrap_recall_means = []
bootstrap_f1_means = []

# Initialize a list to store probability predictions from each bootstrap model for ensemble
ensemble_predictions_proba = []

num_bootstrap_iterations = 100
num_inner_cv_folds = 3

print(f"\nStarting {num_bootstrap_iterations} bootstrap iterations with {num_inner_cv_folds}-fold inner CV and ensemble model training...")

# Get the number of positive participants (all of them will be used)
target_positive_count = len(remaining_participant_ids_class_1)

for i in range(num_bootstrap_iterations):
    # Set random seed for reproducibility of sampling in each bootstrap iteration
    np.random.seed(i)

    # Randomly sample negative class participant IDs with replacement to match positive class count
    sampled_ids_class_0_bootstrap = np.random.choice(remaining_participant_ids_class_0,
                                                     size=target_positive_count,
                                                     replace=True)

    # Combine all positive participant IDs with the sampled negative participant IDs
    balanced_participant_ids_bootstrap = np.concatenate((remaining_participant_ids_class_1,
                                                         sampled_ids_class_0_bootstrap))

    # Filter the main DataFrame (`df_remaining_for_ensemble_training`) to create the bootstrapped dataset
    df_bootstrap_iteration = df_remaining_for_ensemble_training[
        df_remaining_for_ensemble_training['participant_id'].isin(balanced_participant_ids_bootstrap)
    ].copy()

    # Separate features (using top 52) and target from the bootstrapped dataset
    X_bootstrap = df_bootstrap_iteration[top_52_features_suicide]
    y_bootstrap = df_bootstrap_iteration['suicidal_ideation_target']

    # Initialize lists to store metrics for the inner 3-fold CV
    inner_roc_auc_scores = []
    inner_accuracy_scores = []
    inner_precision_scores = []
    inner_recall_scores = []
    inner_f1_scores = []

    # Initialize StratifiedKFold for the inner CV
    skf_inner = StratifiedKFold(n_splits=num_inner_cv_folds, shuffle=True, random_state=42)

    # Perform inner 3-fold CV
    for fold_inner, (train_idx_inner, val_idx_inner) in enumerate(skf_inner.split(X_bootstrap, y_bootstrap)):
        X_train_inner, X_val_inner = X_bootstrap.iloc[train_idx_inner], X_bootstrap.iloc[val_idx_inner]
        y_train_inner, y_val_inner = y_bootstrap.iloc[train_idx_inner], y_bootstrap.iloc[val_idx_inner]

        xgb_classifier_inner = XGBClassifier(eval_metric='logloss', random_state=42)
        xgb_classifier_inner.fit(X_train_inner, y_train_inner)

        y_pred_inner = xgb_classifier_inner.predict(X_val_inner)
        y_pred_proba_inner = xgb_classifier_inner.predict_proba(X_val_inner)[:, 1]

        inner_roc_auc_scores.append(roc_auc_score(y_val_inner, y_pred_proba_inner))
        inner_accuracy_scores.append(accuracy_score(y_val_inner, y_pred_inner))
        inner_precision_scores.append(precision_score(y_val_inner, y_pred_inner, zero_division=0))
        inner_recall_scores.append(recall_score(y_val_inner, y_pred_inner, zero_division=0))
        inner_f1_scores.append(f1_score(y_val_inner, y_pred_inner, zero_division=0))

    # Calculate the mean of metrics from the inner CV for the current bootstrap iteration
    bootstrap_roc_auc_means.append(np.mean(inner_roc_auc_scores))
    bootstrap_accuracy_means.append(np.mean(inner_accuracy_scores))
    bootstrap_precision_means.append(np.mean(inner_precision_scores))
    bootstrap_recall_means.append(np.mean(inner_recall_scores))
    bootstrap_f1_means.append(np.mean(inner_f1_scores))

    # Train a separate model on the entire X_bootstrap and y_bootstrap for the ensemble
    xgb_model_single = XGBClassifier(eval_metric='logloss', random_state=42)
    xgb_model_single.fit(X_bootstrap, y_bootstrap)

    # Make probability predictions on the held-out test set
    y_pred_proba_single = xgb_model_single.predict_proba(X_held_out_test)[:, 1]
    ensemble_predictions_proba.append(y_pred_proba_single)

    if (i + 1) % 10 == 0:
        print(f"Bootstrap Iteration {i+1}/{num_bootstrap_iterations} complete (Inner CV and Ensemble Model Trained).")

print("\n--- Summary of Bootstrap with Inner CV Results (Median and IQR across iterations) ---")
# Calculate the median and interquartile range (IQR) across all bootstrap iterations for inner CV
median_roc_auc_final = np.median(bootstrap_roc_auc_means)
iqr_roc_auc_final = np.percentile(bootstrap_roc_auc_means, 75) - np.percentile(bootstrap_roc_auc_means, 25)

median_accuracy_final = np.median(bootstrap_accuracy_means)
iqr_accuracy_final = np.percentile(bootstrap_accuracy_means, 75) - np.percentile(bootstrap_accuracy_means, 25)

median_precision_final = np.median(bootstrap_precision_means)
iqr_precision_final = np.percentile(bootstrap_precision_means, 75) - np.percentile(bootstrap_precision_means, 25)

median_recall_final = np.median(bootstrap_recall_means)
iqr_recall_final = np.percentile(bootstrap_recall_means, 75) - np.percentile(bootstrap_recall_means, 25)

median_f1_final = np.median(bootstrap_f1_means)
iqr_f1_final = np.percentile(bootstrap_f1_means, 75) - np.percentile(bootstrap_f1_means, 25)

print(f"Median ROC AUC (Inner CV):   {median_roc_auc_final:.4f} (IQR: {iqr_roc_auc_final:.4f})")
print(f"Median Accuracy (Inner CV):  {median_accuracy_final:.4f} (IQR: {iqr_accuracy_final:.4f})")
print(f"Median Precision (Inner CV): {median_precision_final:.4f} (IQR: {iqr_precision_final:.4f})")
print(f"Median Recall (Inner CV):    {median_recall_final:.4f} (IQR: {iqr_recall_final:.4f})")
print(f"Median F1-Score (Inner CV):  {median_f1_final:.4f} (IQR: {iqr_f1_final:.4f})")

print("\n--- Ensemble Model Evaluation on Held-Out Test Set ---")
# Combine and evaluate ensemble model predictions on the held-out test set
ensemble_predictions_proba_array = np.array(ensemble_predictions_proba)
y_pred_proba_ensemble = np.mean(ensemble_predictions_proba_array, axis=0)
y_pred_ensemble = (y_pred_proba_ensemble > 0.5).astype(int)

ensemble_roc_auc = roc_auc_score(y_held_out_test, y_pred_proba_ensemble)
ensemble_accuracy = accuracy_score(y_held_out_test, y_pred_ensemble)
ensemble_precision = precision_score(y_held_out_test, y_pred_ensemble, zero_division=0)
ensemble_recall = recall_score(y_held_out_test, y_pred_ensemble, zero_division=0)
ensemble_f1_score = f1_score(y_held_out_test, y_pred_ensemble, zero_division=0)

print(f"ROC AUC (Ensemble Held-Out):   {ensemble_roc_auc:.4f}")
print(f"Accuracy (Ensemble Held-Out):  {ensemble_accuracy:.4f}")
print(f"Precision (Ensemble Held-Out): {ensemble_precision:.4f}")
print(f"Recall (Ensemble Held-Out):    {ensemble_recall:.4f}")
print(f"F1-Score (Ensemble Held-Out):  {ensemble_f1_score:.4f}")
```
